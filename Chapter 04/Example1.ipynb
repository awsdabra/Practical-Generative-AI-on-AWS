Using Bedrock APIs to list all foundation models

```python
import boto3

# Let's use Amazon Bedrock
bedrock = boto3.client(service_name='bedrock')

response = bedrock.list_foundation_models()
print (response)
```

Using Bedrock API to generate text

```python
import json
bedrockruntime = boto3.client(service_name='bedrock-runtime')

body = json.dumps({
    "max_tokens": 1000,
    "anthropic_version": "bedrock-2023-05-31",
    "messages": [
      {
        "role": "user",
        "content": [
          {
            "type": "text",
            "text": "Explain LLMs to a 8th grader."
          }
        ]
      }
    ]
  })

modelId = 'anthropic.claude-3-haiku-20240307-v1:0'
accept = 'application/json'
contentType = 'application/json'

response = bedrockruntime.invoke_model(body=body, modelId=modelId, accept=accept, contentType=contentType)
response_body = json.loads(response.get('body').read())

# response text
print(response_body.get('content')[0]['text'])
```
